### **Tarea 2d. Valores nulos - conteo por dispositivo**

A continuación, verifique las dos poblaciones con respecto a la variable `device`.

**Pregunta:** ¿Cuántos usuarios de iPhone tenían valores nulos y cuántos usuarios de Android tenían valores nulos?

No hay nada que sugiera una causa no aleatoria de los datos faltantes.


Examine los conteos y porcentajes de usuarios que abandonaron frente a los que se retuvieron. ¿Cuántos de cada grupo están representados en los datos?


```python
# Calcular conteos de usuarios que abandonaron vs. retenidos
print(df['label'].value_counts())
print()
print(df['label'].value_counts(normalize=True))
```

Este conjunto de datos contiene 82% de usuarios retenidos y 18% de usuarios que abandonaron.

A continuación, compare las medianas de cada variable para usuarios que abandonaron y usuarios retenidos. La razón para calcular la mediana y no la media es que no se desea que los valores atípicos afecten indebidamente la descripción de un usuario típico. Observe, por ejemplo, que el valor máximo en la columna `driven_km_drives` es 21,183 km. ¡Eso es más de la mitad de la circunferencia de la Tierra!


```python
# Calcular las medianas de todas las columnas para usuarios que abandonaron y usuarios retenidos
df.groupby('label').median(numeric_only=True)
```

Esto ofrece una instantánea interesante de los dos grupos, usuarios que abandonaron vs. usuarios retenidos:

Los usuarios que abandonaron tuvieron en promedio ~3 viajes más en el último mes que los usuarios retenidos, pero los usuarios retenidos usaron la aplicación en más del doble de días que los usuarios que abandonaron en el mismo periodo.

El usuario mediano que abandonó condujo ~200 km más y 2.5 horas más durante el último mes que el usuario mediano retenido.

Parece que los usuarios que abandonaron hicieron más viajes en menos días, y sus viajes fueron más largos y de mayor duración. Quizás esto sugiera un perfil de usuario. ¡Continúe explorando!



Calcule la mediana de kilómetros por viaje en el último mes para usuarios retenidos y usuarios que abandonaron.

Comience dividiendo la columna `driven_km_drives` entre `drives`. Luego agrupe los resultados por `label` y calcule la mediana de km/viaje de cada grupo.


```python
# Añadir columna `km_per_drive` al df
df['km_per_drive'] = df['driven_km_drives'] / df['drives']

# Agrupar por `label`, calcular la mediana y aislar km por viaje
median_km_per_drive = df.groupby('label').median(numeric_only=True)[['km_per_drive']]
median_km_per_drive
```

El usuario mediano retenido condujo aproximadamente un kilómetro más por viaje que el usuario mediano que abandonó. ¿Cuántos kilómetros por día de conducción fue esto?

Para calcular esta estadística, repita los pasos anteriores usando `driving_days` en lugar de `drives`.


```python
# Añadir columna `km_per_driving_day` al df
df['km_per_driving_day'] = df['driven_km_drives'] / df['driving_days']

# Agrupar por `label`, calcular la mediana y aislar km por día de conducción
median_km_per_driving_day = df.groupby('label').median(numeric_only=True)[['km_per_driving_day']]
median_km_per_driving_day
```

Ahora calcule la mediana del número de viajes por día de conducción para cada grupo.


```python
# Añadir columna `drives_per_driving_day` al df
df['drives_per_driving_day'] = df['drives'] / df['driving_days']

# Agrupar por `label`, calcular la mediana y aislar viajes por día de conducción
median_drives_per_driving_day = df.groupby('label').median(numeric_only=True)[['drives_per_driving_day']]
median_drives_per_driving_day
```

El usuario mediano que abandonó condujo 698 kilómetros cada día que condujo el último mes, lo que equivale a ~240% de la distancia por día de conducción de los usuarios retenidos. El usuario mediano que abandonó también tuvo un número desproporcionado de viajes por día de conducción en comparación con los usuarios retenidos.

Está claro por estas cifras que, independientemente de si un usuario abandonó o no, los usuarios representados en estos datos son conductores intensivos. Probablemente sería razonable suponer que estos datos no representan a los conductores típicos en general. Quizás los datos —y en particular la muestra de usuarios que abandonaron— contienen una alta proporción de conductores de larga distancia.

En consideración de cuánto conducen estos usuarios, valdría la pena recomendar a Waze que recopile más datos sobre estos ‘súper conductores’. Es posible que la razón por la que conducen tanto sea también la razón por la cual la aplicación Waze no satisface sus necesidades específicas, que podrían diferir de las necesidades de un conductor más típico, como un conductor que se desplaza diariamente al trabajo.


Finalmente, examine si existe un desequilibrio en cuántos usuarios abandonaron según el tipo de dispositivo.

Comience obteniendo los conteos totales de cada tipo de dispositivo para cada grupo, usuarios que abandonaron y usuarios retenidos.


```python
# Para cada label, calcular el número de usuarios Android y iPhone
df.groupby(['label', 'device']).size()
```

Ahora, dentro de cada grupo, usuarios que abandonaron y retenidos, calcule qué porcentaje era Android y qué porcentaje era iPhone.


```python
# Para cada label, calcular el porcentaje de usuarios Android y iPhone
df.groupby('label')['device'].value_counts(normalize=True)
```
